```python
import pandas as pd
import numpy as np
```

```python
skyline = pd.read_csv("../lesson-1-2-types-of-data/skyline_enrollments.csv")
skyline.head(10)
```

Nominal columns: count, unique count, mode.
Ordinal columns: mode, median, count.
Interval columns: mean, median, mode, standard deviation, IQR.
Ratio columns: mean, median, mode, standard deviation, IQR, p95.

```python
# Ratio data

hours_studied_mean = skyline["hours_studied"].mean()
hours_studied_mode = skyline["hours_studied"].mode()
hours_studied_median = skyline["hours_studied"].median()
hours_studied_variance = skyline["hours_studied"].var()
hours_studied_std = skyline["hours_studied"].std()
hours_p90 = skyline["hours_studied"].quantile(0.90)
hours_p95 = skyline["hours_studied"].quantile(0.95)
hours_p99 = skyline["hours_studied"].quantile(0.99)

print(f"Mean total hours studied: {hours_studied_mean:.0f}")
print(f"Mode total hours studied: {hours_studied_mode.values}")
print(f"Median total hours studied: {hours_studied_median:.0f}")
print(f"Variance total hours studied: {hours_studied_variance:.0f}")
print(f"Standard Deviation hours studied: {hours_studied_std:.0f}")
print(f"95th percentile: {hours_p95:.0f}")
```

If the mean and median at the same that means that the data is symmetrically distributed. Values are evenly spread around the center with no extreme outliers pulling the average up or down.

```python
# Ordinal data: Final grade

final_grade_mode = skyline["final_grade"].mode()[0]
final_grade_count = skyline["final_grade"].count()

print(f"Final grade mode: {final_grade_mode}")
print(f"Final grade count: {final_grade_count}")

```

```python
# Nominal data: enrollment id, course name, completion status

enrollment_id_mode = skyline["enrollment_id"].mode()[0]
enrollment_id_count = skyline["enrollment_id"].count()
enrollment_id_unique = skyline["enrollment_id"].nunique()

print(f"Enrollment ID mode: {enrollment_id_count}")
print(f"Enrollment ID count: {enrollment_id_mode}")
print(f"Enrollment ID unique count: {enrollment_id_unique}")

course_name_mode = skyline["course_name"].mode()[0]
course_name_count = skyline["course_name"].count()
course_name_unique = skyline["course_name"].nunique()

print(f"Course name mode: {course_name_mode}")
print(f"Course name count: {course_name_count}")
print(f"Course name unique count: {course_name_unique}")

completion_status_mode = skyline["completion_status"].mode()[0]
completion_status_count = skyline["completion_status"].count()
completion_status_unique = skyline["completion_status"].nunique()

print(f"Completion status mode: {completion_status_mode}")
print(f"Completion status count: {completion_status_count}")
print(f"Completion status unique count: {completion_status_unique}")
```

```python
# interval data : enrollment year

enrollment_year_mean   = skyline["enrollment_year"].mean()
enrollment_year_median = skyline["enrollment_year"].median()
enrollment_year_mode   = skyline["enrollment_year"].mode()[0]
enrollment_year_std    = skyline["enrollment_year"].std()
enrollment_year_iqr    = skyline["enrollment_year"].quantile(0.75) - skyline["enrollment_year"].quantile(0.25)

print(f"Enrollment year mean:   {enrollment_year_mean:.0f}")
print(f"Enrollment year median: {enrollment_year_median:.0f}")
print(f"Enrollment year mode:   {enrollment_year_mode}")
print(f"Enrollment year std:    {enrollment_year_std:.0f}")
print(f"Enrollment year IQR:    {enrollment_year_iqr:.0f}")
```
