```python
import pandas as pd
import plotly.express as px

bakery = pd.read_csv("../lesson-1-2-types-of-data/bakery_customers.csv")

fig = px.violin(
    bakery,
    x="region",
    y="total_spent_usd",
    title="Total Spent by Region (Violin Plot)",
    box=True,
    points="all",
)
fig.show()
```
