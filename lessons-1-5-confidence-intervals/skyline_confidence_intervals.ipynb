```python
import numpy as np
import pandas as pd
from scipy import stats
import plotly.express as px
```

```python
students = pd.read_csv("../lesson-1-2-types-of-data/skyline_enrollments.csv")

n = len(students)
sample_mean = students["hours_studied"].mean()
sample_std = students["hours_studied"].std()

print(f"Sample size: n = {n}")
print(f"Sample mean: {sample_mean:.0f}")
print(f"Sample standard deviation: {sample_std:.0f}")
```

```python
# standard error
standard_error = sample_std / np.sqrt(n)
z = 1.96

# Margin of error
margin_of_error = z * standard_error

# Confidence Interval

ci_upper = sample_mean + margin_of_error
ci_lower = sample_mean - margin_of_error

print(f"Standard error: {standard_error:.0f}")
print(f"Margin of error 95%: {margin_of_error:.0f}")
print(f"95% Confidence Interval for the mean: {ci_lower:.0f}, {ci_upper:.0f}")
```

The average hours studied in the sample is 30 hours. Based on the sample size of 100 students and a sample standard deviation of 5, a 95% confidence interval for the true sample mean is 29 to 31. So we can say with 95% confidence that the true average hours studied is somewhere in that range.

```python
# Sample portion. Students who completed the course.

n = len(students)
n_completed = (students["completion_status"] == "Completed").sum()
p_hat = n_completed / n

print(f"Sample size: {n}")
print(f"Students who completed the courses: {n_completed}")
print(f"Sample proportion: {p_hat}")
```

```python
# Standard error of the proportion
se_prop = np.sqrt(p_hat * (1 - p_hat) / n)

# 95% CI
z = 1.96
moe_prop = z * se_prop
ci_prop_lower = p_hat - moe_prop
ci_prop_upper = p_hat + moe_prop

print(f"Standard error: {se_prop:.4f}")
print(f"Margin of error (95%): {moe_prop:.4f}")
print(f"95% CI for students who completed the course: ({ci_prop_lower:.3f}, {ci_prop_upper:.3f})")
print(f"Or: {p_hat:.1%} ± {moe_prop:.1%}")
```

If we resampled and re computed many times, 95% of the resulting intervals would capture the true average hours studied for students enrolled on the courses.
